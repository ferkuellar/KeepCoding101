# (4)

Crea la funcin `sum_mult_of_3`:

1. Recibe una lista de nmeros
2. Devuelve la suma de aquellos que son
3. mltiplos de 3
   
## DIVIDE & VENCERS

Resuelve primero el subproblema `saber si un nmero es mltiplo de 3`, creando el predicado `is_multiple_3`. Recuerda el operador `%` para obtener el resto de una divisin.

```python
def is_multiple_3(number):
    return number % 3 == 0

def sum_mult_of_3(numbers):
    sum = 0
    for number in numbers:
        if is_multiple_3(number):
            sum += number
    return sum

numbers = [5, 7, 8, 11]
# numbers = [3, 6, 9, 12]

result = sum_mult_of_3(numbers)
print(result) 
```
