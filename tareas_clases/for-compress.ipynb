# (1)

Escriba la funcion `sum_all`que recibe una lista de numero y devuelve la suma total.

En este ejercicio, veremos el proceso general que hay que seguir para diseñar una función. Empezaremos con un ejemplo sencillo y lo iremos complicando poco a poco.

Vamos a aplicar el algoritmo universal.

## Entrada y Salida.
    
    `sum_all` devolvera siempre un numero.
    `sum_all` recibira una lista de numeros.

## Casos especiales 

Cuando trabajamos con listas, es importante tener en cuenta que pueden estar vacías. En ese caso, no podemos calcular la suma de sus elementos, porque no tiene ninguno. Por lo tanto, la función `sum_all` debe devolver 0 cuando reciba una lista vacía.

Hay otro caso más: si recibe una lista de cosas que `no son numeros`. En ese caso, no hay mucho quue hacer, la funcion `sum_all` es una funcion parsila y tendremos que gesyionar el error.

## Firma y comentarios

    `def sum_all ( list ):`

        `# list: una lista de numeros`
        `# Devuelve: un numero entero`

en la segunda etapa, la funcion tendria esta pinta:

    `def sum_all (list):`
        `"""`
        `Suma todos los elementos de una lista de numero.`
        `Si recibe la lista vacia, devuelve cero.`
        `"""`
        `# inicializamos la variable total`
        `total = ??`
        `# recorremos la lista, sumando el valor de cada elemento a total.`
        `for list in list:`
            `????`
        `# cuando terminamos, devolvemos el total ( que ya no sera parcial)`
        `return ??`
1.  Finiquita tu ese ejercicio.con los datos que faltan.
2.  Comprueba que funciona tal y como dices, casos especiales incluidos.

```python
def sum_all (numbers):
    total = 0
    for n in numbers:
        total += n
    return total
print(sum_all([1,10,35]))
```

# (2)
Siguiendo el método de arriba, crea la función `multiply_all`, que recibe una lista de números ydevuelve el producto de todos ellos.

```python
def multiply_all (numbers):
    total = 1
    for n in numbers:
        total *= n
    return total
print(multiply_all([2,10,35]))
```

# (4)

Crea la función `sum_mult_of_3`:

1. Recibe una lista de números
2. Devuelve la suma de aquellos que son
3. múltiplos de 3
   
## DIVIDE & VENCERÁS

Resuelve primero el subproblema `saber si un número es múltiplo de 3`, creando el predicado `is_multiple_3`. Recuerda el operador `%` para obtener el resto de una división.

```python
def is_multiple_3(number):
    return number % 3 == 0

def sum_mult_of_3(numbers):
    sum = 0
    for number in numbers:
        if is_multiple_3(number):
            sum += number
    return sum

numbers = [5, 7, 8, 11]
# numbers = [3, 6, 9, 12]

result = sum_mult_of_3(numbers)
print(result) 
```
