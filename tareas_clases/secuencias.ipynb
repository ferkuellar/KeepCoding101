# Buscar en una secuencia 

Para buscar en una secuencia, se utiliza el operador `in`:

         `elemento in secuencia` devuelve `True` si el elemento est en la secuencia y `False` en caso contrario.


```python
vowels = ['a', 'e', 'i', 'o', 'u']
"a" in vowels
"w" in vowels
```
