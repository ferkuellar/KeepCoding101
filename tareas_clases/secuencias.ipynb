# Secuencias en Python

Una secuencia es cualquier estructura de datos que nos permite tenerlos uno detrás de otro.La más sencillaes la lista (hay otras).

        · Las secuencias tienen una longitud (len)
        · Las secuencias se pueden recorrer
        · En la secuencias se puede buscar un elemento


# Listas
Las listas son un tipo de Secuencia.
Las listas en Python se representan como una secuencia de datos separados por comas y entre corchetes:
        · [1, 8, 93]
        · [True]
        · [[5,6], ["hola", "mundo"]]
        · []
        · [True, 45, 'hola']

Las listas pueden ser de cualquier cosa, incluso otras listas o estar vacías.

Cada lista es un valor de Python: se evalúan a sí mismos.

# El tipo de lista

Las listas son un tipo de Python:

        · Tienen infinitas instancias
        · Tienen operaciones en común:
                · tienen longitud (len)
                · se pueden recorrer
                · se pueden buscar
                · se puede seleccionar un elemento
                · se puede reemplazar un elemento

# Buscar en una secuencia 

Para buscar en una secuencia, se utiliza el operador `in`:

        · `elemento in secuencia` devuelve `True` si el elemento está en la secuencia y `False` en caso contrario.


```python
vowels = ['a', 'e', 'i', 'o', 'u']
"a" in vowels
"w" in vowels
```

# Recorrer una secuencia

Es muy comun recorrer una secuencia, es decir, visitar cada uno de sus elementos. Hay varias formas de hacerlo:

        · Con un bucle `for`
        · Con un bucle `while`
        · Con un bucle `for` y la función `range`
